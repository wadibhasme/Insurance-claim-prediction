### Find  out Highly Correlated features 

```python
#### with the following function we can select the highly correlated features by seting threshold value as 0.8
def correlation(df,thresold):
    col_corr=set()   ### set of all the name of correlated columns
    corr_matrix=df.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (abs(corr_matrix.iloc[i,j]))> thresold:
                colname=corr_matrix.columns[i]  ### geting the name of columns
                col_corr.add(colname)
    return  col_corr     
```

```python
corr_features=list(correlation(df_copy,0.8))
#len(set(corr_features))
corr_features
```

```python
def remove_outlier(data):
    points=[]
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if(np.abs(z_score)>threshold):
            points.append(i)
            
    return points

out_pop_density=remove_outlier(df_copy["population_density"])    
```

```python
########### Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
######## KNN
##### Perform Feature scaling to bring the data on same scale
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### KNN 

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, y_train)
```

```python
knn_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import f1_score
knn_f1 = f1_score(y_test,knn_pred )
print(f'The f1 score for K nearest neignbours is {round(knn_f1*100,2)}%')
```

### Random Forest Classifier 

```python
######### Random Forest
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
ran_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import f1_score
ran_f1 = f1_score(y_test,ran_pred)
print(f'The f1 score for random Forest Classifier is {round(ran_f1*100,2)}%')
```

### Logistic Regression 

```python
####### logistic regression
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(X_train, y_train)
```

```python
log_pred = classifier.predict(X_test)
```

```python
log_f1 = f1_score(y_test,log_pred )
print(f'The f1 score for logistic regression is {round(log_f1*100,2)}%')
```

### Naive Bayes Classifier 

```python
######### Naive Bayes
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
Naive_pred = classifier.predict(X_test)
```

```python
Naive_f1 = f1_score(y_test,Naive_pred )
print(f'The f1 score for Naive Bayes is {round(Naive_f1*100,2)}%')
```

### Support vector Machine 

```python
######## SVM
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
comparison = pd.DataFrame({
    "Logistic regression":{'F1 Score':log_f1},
    "K-nearest neighbours":{'F1 Score':knn_f1},
    "random_forest_classifier":{'F1 Score':ran_f1},
    "Support vector machine":{'F1 Score':svm_f1},
    "Naive bayes":{'F1 Score':Naive_f1}
}).T
```
